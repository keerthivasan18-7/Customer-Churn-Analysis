# EDA and Preprocessing
This notebook documents loading the Telco Customer Churn dataset, a short exploratory data analysis (EDA), and the preprocessing steps used to create the processed CSV files in `data/`.

```python
# Imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
pd.options.display.max_columns = 200
pd.options.display.width = 120
```

```python
# Load raw and processed data
raw_path = os.path.join('data', 'WA_Fn-UseC_-Telco-Customer-Churn.csv')
processed_path = os.path.join('data', 'processed_telco.csv')
raw = pd.read_csv(raw_path)
processed = pd.read_csv(processed_path)
print('raw shape:', raw.shape)
print('processed shape:', processed.shape)
```

```python
# Quick preview
display(raw.head())
display(raw.dtypes)
display(processed.head())
display(processed.dtypes)
```

```python
# Missing values and target distribution
print('Missing in raw:')
print(raw.isnull().sum())
print('
Missing in processed:')
print(processed.isnull().sum())
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=raw)
plt.title('Churn distribution (raw)')
plt.show()
```

```python
# Numeric variable exploration: MonthlyCharges and TotalCharges
plt.figure(figsize=(8,4))
sns.histplot(raw['MonthlyCharges'], kde=True, bins=30)
plt.title('MonthlyCharges distribution')
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=raw)
plt.title('MonthlyCharges by Churn')
plt.show()
```

```python
# Tenure groups and churn rate
raw['tenure_group'] = pd.cut(raw['tenure'], bins=[0,12,24,48,999], labels=['0-12','12-24','24-48','48+'])
tg = raw.groupby('tenure_group')['Churn'].apply(lambda x: (x=='Yes').mean())
tg = tg.sort_index()
plt.figure(figsize=(6,4))
tg.plot(kind='bar')
plt.ylabel('Churn rate')
plt.title('Churn rate by tenure group')
plt.show()
display(tg)
```
